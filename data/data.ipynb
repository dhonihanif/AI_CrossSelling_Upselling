```python
trainer = df["trainer"].values.tolist()
trainerr = set(trainer)
trainer2 = [[i, trainer.count(i)] for i in trainerr]
trainer2.sort(key=lambda x:x[1], reverse=True)
choice = [i[0] for i in trainer2]
trainer = [i for i in df["trainer"]] + [np.random.choice(choice, p=[0.3, 0.2, 0.2, 0.1, 0.1] + ([0.1/25] * 25)) for i in range(5000-len(df))]
trainer = pd.Series(trainer)
print(trainer.value_counts())
```

```python
def company_id(num_row):
    list_id = []

    for i in range(num_row):
        num = i + 1
        id = 'P' + '{:04d}'.format(num)
        np.array(list_id.append(id))

    return list_id

def product_id(num_row):
    list_id = []

    for i in range(num_row):
        num = i + 1
        id = 'B' + '{:04d}'.format(num)
        np.array(list_id.append(id))

    return list_id

company_ID = company_id(5000)
training_ID = product_id(5000)
```

```python
fake = Faker("id_ID")

company = [i for i in df["perusahaan"]] + [fake.company() for i in range(5000 - len(df))]
alamat = [fake.address() for i in range(5000)]
telepon = [fake.phone_number() for i in range(5000)]
email = [fake.email() for i in range(5000)]
topik = [i for i in df["topik"]]
pic = [fake.name() for i in range(5000)]
sector = DynamicProvider(
    provider_name="sector",
    elements=["Pertanian, Kehutanan dan Perikanan", "Pertambangan dan Penggalian", "Industri Pengolahan", 
              "Pengadaan Listrik, Gas, Uap/Air Panas Dan Udara Dingin", "Treatment Air, Treatment Air Limbah, Treatment dan Pemulihan Material Sampah, dan Aktivitas Remediasi", 
              "Konstruksi", "Perdagangan Besar Dan Eceran; Reparasi Dan Perawatan Mobil Dan Sepeda Motor", "Pengangkutan dan Pergudangan",
              "IPenyediaan Akomodasi Dan Penyediaan Makan Minum", "Informasi Dan Komunikasi", "Aktivitas Keuangan dan Asuransi",
              "Real Estat", "Aktivitas Profesional, Ilmiah Dan Teknis", "Aktivitas Penyewaan dan Sewa Guna Usaha Tanpa Hak Opsi, Ketenagakerjaan, Agen Perjalanan dan Penunjang Usaha Lainnya",
              "Administrasi Pemerintahan, Pertahanan Dan Jaminan Sosial Wajib", "Pendidikan", "Aktivitas Kesehatan Manusia Dan Aktivitas Sosial",
              "Kesenian, Hiburan Dan Rekreasi", "Aktivitas Jasa Lainnya", "Aktivitas Rumah Tangga Sebagai Pemberi Kerja; Aktivitas Yang Menghasilkan Barang Dan Jasa Oleh Rumah Tangga yang Digunakan untuk Memenuhi Kebutuhan Sendiri",
              "Aktivitas Badan Internasional Dan Badan Ekstra Internasional Lainnya"
]
)
group = DynamicProvider(
    provider_name="group_retail",
    elements=["group", "retail"]
)
fake.add_provider(sector)
fake.add_provider(group)

sectors = [fake.sector() for i in range(5000)]
groups = [fake.group_retail() for i in range(5000)]

df2 = pd.DataFrame()
df2["company_ID"] = company_ID
df2["company_name"] = company
df2["company_pic"] = pic
df2["company_address"] = alamat
df2["company_phone"] = telepon
df2["company_email"] = email
df2["company_field"] = sectors
df2["transaction_date"] = pd.date_range("20200101", periods=5000)
df2["training_ID"] = training_ID
df2.head()
```
