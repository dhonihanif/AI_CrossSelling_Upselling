```python
def generate_discount(number_of_participants):
    if number_of_participants >= 12:
        return np.random.choice([40, 45, 50, 55, 60], p=[0.5, 0.1, 0.2, 0.1, 0.1]), "Group"
    elif number_of_participants >= 7:
        return np.random.choice([20, 25, 30, 35], p=[0.4, 0.2, 0.2, 0.2]), "Group"
    elif number_of_participants > 2:
        return np.random.choice([5, 10, 15], p=[0.6, 0.3, 0.1]), "Retail"
    else:
        return np.random.choice([0, 2], p=[0.9, 0.1]), "Retail"
    
for i in range(len(df)):
    dskn, grp = generate_discount(df.loc[i, "number_of_participants"])
    group.append(grp)
    diskon.append(dskn)
```
