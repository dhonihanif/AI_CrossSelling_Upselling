```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./AI Sales Cross Selling and UpSelling.csv")
df.head()
```

```python
df.drop(["Unnamed: 0", "prefered_category", "training_type", "training_name", "cross_selling", "upselling"], axis=1, inplace=True)
df.head()
```

```python
collect = []
cross_selling = []
upselling = []
data_sales = [
    "Mengelola NPL dengan Early Warning System (EWS)",
    "Manajemen Resiko APU & PPT bagi Staff Perbankan, Multifinance, sekuritas dan asuransi",
    "Graphology & Powerful Interview for Credit Analyst",
    "Restrukturisasi Kredit for SME Lending",
    "Debt Collection Strategy & Tactics",
    "Teknik Penyelesaian Kredit & Pemasaran AYDA yang Efektif",
    "How to detect & prevent application fraud in Digital Era for Banking & Fintech",
    "Managing Corporate Collection"
]
for i in range(len(df)):
    if df.loc[i, "company_field"] in ["Pertanian, Kehutanan dan Perikanan",
                                      "Pengadaan Listrik, Gas, Uap/Air Panas Dan Udara Dingin", "Treatment Air, Treatment Air Limbah, Treatment dan Pemulihan Material Sampah, dan Aktivitas Remediasi", 
                                      "Aktivitas Kesehatan Manusia Dan Aktivitas Sosial"]:
        collect.append(np.random.choice(data_sales, p=[0.1/7]*7 + [0.9]))
        cross_selling.append(np.random.choice(data_sales, p=[0.1/7] + [0.9] + [0.1/7]*6))
        upselling.append(np.random.choice(data_sales, p=[0.1/7]*2 + [0.9] + [0.1/7]*5))
    elif df.loc[i, "company_field"] in ["Kontruksi", "Perdagangan Besar Dan Eceran; Reparasi Dan Perawatan Mobil Dan Sepeda Motor",
                                        "Informasi Dan Komunikasi", "Pendidikan", "Aktivitas Profesional, Ilmiah Dan Teknis",
                                        "Kesenian, Hiburan Dan Rekreasi"
                                        ]:
        collect.append(np.random.choice(data_sales, p=[0.1/7]*2 + [0.9] + [0.1/7]*5))
        cross_selling.append(np.random.choice(data_sales, p=[0.1/7]*3 + [0.9] + [0.1/7]*4))
        upselling.append(np.random.choice(data_sales, p=[0.1/7]*6 + [0.9] + [0.1/7]))
    elif df.loc[i, "company_field"] in ["Aktivitas Keuangan dan Asuransi", "Real Estat",
                                        "Aktivitas Penyewaan dan Sewa Guna Usaha Tanpa Hak Opsi, Ketenagakerjaan, Agen Perjalanan dan Penunjang Usaha Lainnya",
                                        ]:
        collect.append(np.random.choice(data_sales, p=[0.9] + [0.1/7]*7))
        cross_selling.append(np.random.choice(data_sales, p=[0.1/7]*7 + [0.9]))
        upselling.append(np.random.choice(data_sales, p=[0.1/7] + [0.9] + [0.1/7]*6))
    elif df.loc[i, "company_field"] in ["Administrasi Pemerintahan, Pertahanan Dan Jaminan Sosial Wajib",
                                        "Pertambangan dan Penggalian"]:
        collect.append(np.random.choice(data_sales, p=[0.1/7] + [0.9] + [0.1/7]*6))
        cross_selling.append(np.random.choice(data_sales, p=[0.1/7]*4 + [0.9] + [0.1/7]*3))
        upselling.append(np.random.choice(data_sales, p=[0.1/7]*5 + [0.9] + [0.1/7]*2))
    else:
        collect.append(np.random.choice(data_sales, p=[0.1/7]*6 + [0.9] + [0.1/7]))
        cross_selling.append(np.random.choice(data_sales, p=[0.1/7]*6 + [0.9] + [0.1/7]))
        upselling.append(np.random.choice(data_sales, p=[0.1/7]*7 + [0.9]))


```

```python
df["training_type"] = "Credit & Collection Series"
df["training_name"] = collect
df["cross_selling"] = cross_selling
df["upselling"] = upselling
df.head()
```

```python
df.to_csv("./dataset/Credit & Collection Series.csv")
```
